## Compare The Triplets
**Platform: HackerRank**
#### Problem: Taking input -> int a[3]: Alice's challenge rating, int b[3]: Bob's challenge rating
#### Output: int[2]: the first element is Alice's score and the second is Bob's score

```python
def compareTriplets(a, b):
    result=[0,0]
    for i, j in zip(a, b):
        if i>j:
            result[0] += 1
        elif i<j:
            result[1] += 1
        else:
            None
    return result
        
if __name__ == '__main__':
    a = list(map(int, input().split()))

    b = list(map(int, input().split()))

    result = compareTriplets(a, b)
    print(result)
```

### My Approach:
Using zip() function to combine two lists/arrays and then iteration and performing comparision
if a[i]>b[i] then Alice gets 1 point
if a[i]< b[i] then Bob gets 1 point
if a[i]=b[i] then no point to both of them

Using result array at index 0 stores Alice's score and at index 1 stores Bob's score

### Time Complexity: O(n)
here n is size of array
### Space Complexity: O(1)
as result array is constant throughout the program

### Sample Input 
99 16 8

### Sample Output
2 1
