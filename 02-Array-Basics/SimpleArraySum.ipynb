## Simple Array Sum
**Platform: HackerRank**
#### Problem: Taking input -> n: size of array, arr: element of array in one line space seperated
#### Output: sum of elements in array

```python
def simpleArraySum(arr):
    total = 0
    for i in arr:
        total += i
    return total
if __name__ == "__main__":
    n = int(input("Enter total number of elements of array"))
    arr = list(map(int,input("enter elements seperated by space").split(" ")))
    print(simpleArraySum(arr))
```

### My Approach:
Using loop and a sum variable for storing sum of elements in array
the i contains element of array which get added in sum' value one by one till the i reach to the last element of the array

### Time Complexity: O(n)
here n is size of array, as we have to iterate one by one each element and add them 
### Space Complexity: O(1)
as here loop variable i and sum use constant memory space 

### Sample Input
Enter total number of elements of array 5

enter elements seperated by space 1 2 3 4 5

### Sample Output
15
