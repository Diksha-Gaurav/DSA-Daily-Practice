## Second Largest Element in the Array
#### Problem: Taking input -> an array of n elements
#### Output: 2nd largest element 

### Method 1 - Brute Force Approach
Using build in sort() method of list that will sort the array in increasing order. Then accessing 2nd last element using backward index a[-2] which give us 2nd largest element.

### Time Complexity: O(n log n)
### Space Complexity: O(1)

```python
a =[ 55,32,97, -55,45,32,88,21]
a.sort()
print(a[-2])
```

### Method 2 - Better Approach
Using loop, the 1st loop find the largest element and the 2nd loop will find second largest element by comparing ith element > second_largest element and also the ith element should not be equal to the largest one.

### Time Complexity:O(2n) ~= O(n)
### Space Complexity: O(1)

```python
a =[ 55,32,97,-55,45,32,88,21]
largest = float('-inf')
second_largest = float('-inf')
for i in range(0,len(a)):
    largest = max(largest,a[i])
for i in range(0,len(a)):
    if (second_largest < a[i]) and (a[i] != largest):
        second_largest = a[i]
print(second_largest)
```

### Method 3 - Optimal Approach
The largest and second_largest are initialized with the smallest possible value and now only single loop checks each element of the list one by one
If the ith element is greater than largest, largest is updated and the old largest becomes second_largest Or if the element is smaller than largest but greater than second_largest, second_largest is updated.

### Time Complexity: O(n)
### Space Complexity: O(1)

```python
a =[ 55,32,97,-55,45,32,88,21]
largest = float('-inf')
second_largest = float('-inf')
for i in range (0,len(a)):
    if a[i]> largest:
        second_largest= largest
        largest = a[i]
    elif (second_largest < a[i]) and (a[i] != largest):
        second_largest = a[i]
    
print(second_largest)

```
