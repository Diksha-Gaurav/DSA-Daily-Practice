## Largest Element in the Array
#### Problem: Taking input -> an array of n elements
#### Output: largest element 

```python
def largest(arr):
    largest= arr[0] ## also write it as largest = float("-inf")
    for i in range (1,(len(arr))):
        largest = max(largest, arr[i])
    return largest

if __name__ == "__main__":
    n = int(input("Enter size of array"))
    ar = list(map(int,input("Enter elements").rstrip().split()))
    max_element = largest(ar)
    print(max_element)
```

### My Approach:
A largest variable stores first element of array as largest one. Then loop starts for index 1 and find maximum element by comparing largest one and ith element of array:
if largest value > array[i] then no overwrite
if largest value < array[i] then largest value get overwrite

### Time Complexity: O(n)
here n is size of array
### Space Complexity: O(1)
as largest variable use constant memory space

### Sample Input 
Enter size of array 6
Enter elements 55 32 -97 99 3 67

### Sample Output
99
