# COUNTING OF DIGITS
### Method 1: using while loop
Loop run until number is not equal to zero 0
we count digit and remove last digit using floor division //

## Time Complexity :
whenever your loop's iteration depends upon division by specific number then time complexity = log base 10(N)
here base is 10 because we are dividing by 10; in other case it can be any number from which division is performed

## Space Complexity:
O(1)
as no extra space is requires for storage

```python
num = int(input("enter the number to find count of digits"))
temp = num
count = 0
while num>0:
    count = count + 1
    num = num // 10

print(f"The number of digits in {temp} is {count}")
```

### Method 2: using log function maths library


```python
from math import *
def count_digits(num):
    count = int(log10(num) + 1)
    return count

num = int(input("enter a number"))
count_digits(num)
```

# PALINDROME OF NUMBER
Number whose reverse is same as of its original
### TIME COMPLEXITY:
O(log 10(N))
### SPACE COMPLEXITY:
O(1)

```python
def Palindrome(num):
    temp = num
    reverse = 0
    if num < 0:
        return False
    while num != 0:
        remainder = num % 10
        reverse = reverse * 10 + remainder
        num = num // 10
    if temp == reverse:
        print("Number is palindrome")

    else:   
        print("Number is not palindrome")
        
num = int(input("enter the number to check palindrome"))
Palindrome(num)
```

### O[n * m ]

```python
l = [[4,5],[6,7,8,1]]
sum = 0
for i in l:
    for j in i:
        sum = sum + j
print(sum)
```
